```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data from Excel file
def load_data_from_excel(file_path):
    data = pd.read_excel(file_path, parse_dates=['date'], index_col='date', engine='openpyxl')
    
    # Ensure data is sorted by date
    data = data.sort_index()
    return data

# Moving Average Crossover with EMA Strategy
def moving_average_crossover_ema(data, short_window=5, long_window=20):
    data['short_ema'] = data['close'].ewm(span=short_window, adjust=False).mean()
    data['long_ema'] = data['close'].ewm(span=long_window, adjust=False).mean()
    
    # Buy signal: short EMA crosses above long EMA
    data['signal'] = np.where(data['short_ema'] > data['long_ema'], 1, 0)
    
    # Generate signals (1 for buy, -1 for sell)
    data['position'] = data['signal'].diff()
    
    return data

# VWAP Scalping Strategy
def vwap(data):
    data['cum_vol'] = data['volume'].cumsum()
    data['cum_vol_price'] = (data['close'] * data['volume']).cumsum()
    data['vwap'] = data['cum_vol_price'] / data['cum_vol']
    
    return data

# Plotting function to visualize the strategies
def plot_signals(data, ticker):
    plt.figure(figsize=(12, 8))
    
    # Plot closing price and moving averages (EMA)
    plt.subplot(2, 1, 1)
    plt.plot(data['close'], label=f'{ticker} Close')
    plt.plot(data['short_ema'], label='Short-term EMA (5)', linestyle='--')
    plt.plot(data['long_ema'], label='Long-term EMA (20)', linestyle='--')
    
    # Mark buy and sell signals for EMA Crossover
    buy_signals = data[data['position'] == 1]
    sell_signals = data[data['position'] == -1]
    plt.scatter(buy_signals.index, buy_signals['close'], marker='^', color='g', label='Buy Signal')
    plt.scatter(sell_signals.index, sell_signals['close'], marker='v', color='r', label='Sell Signal')
    
    plt.title(f'{ticker} Moving Average Crossover (EMA) Strategy')
    plt.legend()
    
    # Plot VWAP
    plt.subplot(2, 1, 2)
    plt.plot(data['close'], label=f'{ticker} Close')
    plt.plot(data['vwap'], label='VWAP', color='orange')
    
    plt.title(f'{ticker} VWAP')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Main function to execute strategies and calculate balance
def main():
    # Provide the path to your data.xlsx file
    file_path = 'n.xlsx'
    
    # Load data from Excel file
    data = load_data_from_excel(file_path)
    
    # Ensure the required columns are present
    required_columns = ['date', 'open', 'high', 'low', 'close', 'volume']
    # if not all(col in data.columns for col in required_columns):
    #     print(f"Error: The Excel file must contain columns: {', '.join(required_columns)}")
    #     return
    
    # Calculate Daily Return (percentage change)
    data['daily_return'] = data['close'].pct_change()  # Calculate daily return as percentage change in close price
    
    # Apply Moving Average Crossover Strategy (EMA)
    data = moving_average_crossover_ema(data)
    
    # Apply VWAP Scalping Strategy
    data = vwap(data)
    
    # Initial balance (assumed ₹1,000,000)
    initial_balance = 1000000
    balance = initial_balance
    trade_history = []
    stop_loss_pct = 0.05  # 2% Stop Loss
    take_profit_pct = 0.05  # 5% Take Profit

    # Simulate trading with Buy/Sell signals
    for i in range(1, len(data)):
        # Apply entry confirmation: price must be above VWAP and EMA crossover signal must be 1 (buy)
        if data['position'].iloc[i] == 1 and data['close'].iloc[i] > data['vwap'].iloc[i]:  # Buy signal confirmation
            buy_price = data['close'].iloc[i]
            stop_loss = buy_price * (1 - stop_loss_pct)
            take_profit = buy_price * (1 + take_profit_pct)
            trade_history.append({'trade': 'buy', 'price': buy_price, 'stop_loss': stop_loss, 'take_profit': take_profit})
            print(f"Buying at {buy_price} on {data.index[i]}")
        
        elif data['position'].iloc[i] == -1:  # Sell signal (EMA)
            # Check if price reached stop loss or take profit before selling
            last_trade = trade_history[-1]
            sell_price = data['close'].iloc[i]
            
            # If price hits stop loss or take profit, sell
            if sell_price <= last_trade['stop_loss'] or sell_price >= last_trade['take_profit']:
                trade_profit = sell_price - last_trade['price']
                balance += trade_profit  # Update balance with profit or loss from trade
                trade_history.append({'trade': 'sell', 'price': sell_price, 'profit': trade_profit})
                print(f"Selling at {sell_price} on {data.index[i]}, Profit/Loss: {trade_profit:.2f}")
    
    # Final balance and profit/loss
    print(f"\nInitial Balance: ₹{initial_balance:.2f}")
    print(f"Final Balance: ₹{balance:.2f}")
    total_profit = balance - initial_balance
    print(f"Total Profit/Loss: ₹{total_profit:.2f}")
    
    # Performance of VWAP Scalping (Buy when above VWAP, Sell when below)
    vwap_returns = data[data['close'] > data['vwap']]['daily_return'].sum()  # Buy signal above VWAP
    vwap_loss = data[data['close'] < data['vwap']]['daily_return'].sum()  # Sell signal below VWAP
    print(f'VWAP Scalping Strategy: {vwap_returns:.2%} returns, {vwap_loss:.2%} losses')
    
    # Visualize results
    plot_signals(data, 'Nifty')
    
if __name__ == "__main__":
    main()
```
