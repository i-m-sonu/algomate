```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import backtrader as bt

# Ensure Matplotlib works in Jupyter Notebook
import matplotlib
matplotlib.use('Agg')  # Use a non-interactive backend
# %matplotlib inline  # Force inline plotting

# Load historical data from an Excel file
file_path = 'data.xlsx'  # Make sure this file is in the working directory
df = pd.read_excel(file_path)

# Convert date column to datetime and set it as index
df['Date'] = pd.to_datetime(df['date'])  # Ensure Date column is in datetime format
df.set_index('Date', inplace=True)

# Ensure Backtrader-compatible column names
df.rename(columns={'Open': 'open', 'High': 'high', 'Low': 'low', 'Close': 'close', 'Volume': 'volume'}, inplace=True)

# Define the combined scalping strategy class
class CombinedScalpingStrategy(bt.Strategy):
    params = (
        ('short_ema', 9), 
        ('long_ema', 21), 
        ('rsi_period', 14), 
        ('bollinger_period', 20), 
        ('bollinger_dev', 2),
        ('macd_short', 12), 
        ('macd_long', 26), 
        ('macd_signal', 9),
    )

    def __init__(self):
        self.short_ema = bt.indicators.EMA(self.data.close, period=self.params.short_ema)
        self.long_ema = bt.indicators.EMA(self.data.close, period=self.params.long_ema)
        self.rsi = bt.indicators.RSI(period=self.params.rsi_period)
        self.bollinger = bt.indicators.BollingerBands(period=self.params.bollinger_period, devfactor=self.params.bollinger_dev)
        self.macd = bt.indicators.MACD(self.data.close, period_me1=self.params.macd_short, 
                                       period_me2=self.params.macd_long, period_signal=self.params.macd_signal)

        self.initial_cash = self.broker.get_cash()  # Store the initial capital
        self.trade_log = []  # Store trade history
    
    def log(self, txt):
        dt = self.datas[0].datetime.datetime(0)
        print(f'{dt} - {txt}')

    def notify_trade(self, trade):
        if trade.isclosed:
            profit = trade.pnl  # Profit/Loss from the trade
            self.log(f'TRADE CLOSED - Profit/Loss: ${profit:.2f}')
            self.trade_log.append((trade.dtclose, profit))

    def next(self):
        if self.short_ema > self.long_ema and self.rsi < 30 and self.macd.macd > self.macd.signal:
            if not self.position:
                self.buy()
                self.log(f'BUY ORDER PLACED - Price: ${self.data.close[0]:.2f}')

        elif self.short_ema < self.long_ema and self.rsi > 70 and self.macd.macd < self.macd.signal:
            if self.position:
                self.sell()
                self.log(f'SELL ORDER PLACED - Price: ${self.data.close[0]:.2f}')

    def stop(self):
        final_cash = self.broker.get_cash()  # Store the final capital
        total_profit_loss = final_cash - self.initial_cash
        print("\n===== BACKTEST RESULTS =====")
        print(f"Initial Capital: ${self.initial_cash:.2f}")
        print(f"Final Capital:   ${final_cash:.2f}")
        print(f"Total Profit/Loss: ${total_profit_loss:.2f}")
        print("===========================\n")

# Convert Pandas DataFrame to Backtrader-compatible data feed
data_feed = bt.feeds.PandasData(dataname=df)

# Create backtest environment
cerebro = bt.Cerebro()
cerebro.adddata(data_feed)
cerebro.addstrategy(CombinedScalpingStrategy)
cerebro.broker.set_cash(10000)  # Set initial capital
cerebro.broker.setcommission(commission=0.001)  # FIXED: Use setcommission()
cerebro.addsizer(bt.sizers.FixedSize, stake=10)

# Run the backtest
cerebro.run()

# Save the plot instead of displaying it directly
fig = cerebro.plot(style='candlestick')[0][0]  # Get the figure
fig.savefig("backtest_results.png")  # Save as an image

# Display the saved image
from IPython.display import display
from PIL import Image
display(Image.open("backtest_results.png"))
```
