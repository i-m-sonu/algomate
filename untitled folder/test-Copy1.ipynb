```python
import pandas as pd
import numpy as np
import talib
import matplotlib.pyplot as plt
from datetime import time

# Configuration
pd.set_option('display.max_columns', None)
plt.style.use('ggplot')

class IntradayBacktester:
    def __init__(self, filepath='data.xlsx'):
        self.initial_capital = 1_000_000
        self.position_size = 0.95
        self.transaction_cost = 0.0005
        self.slippage = 0.0002
        self.trading_session_start = time(9, 15)
        self.trading_session_end = time(15, 30)

        try:
            self.df = self.load_data(filepath)
            print("Data loaded successfully. Shape:", self.df.shape)
        except FileNotFoundError:
            print(f"Error: File '{filepath}' not found.")
            exit()
        except KeyError as e:
            print(f"Error: {e}")
            exit()
        except Exception as e:
            print(f"Unexpected error: {e}")
            exit()

    def load_data(self, filepath):
        """Load and preprocess 1-minute OHLCV data"""
        df = pd.read_excel(filepath)
        
        # Convert 'date' column to datetime with explicit format
        df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y %H:%M')
        df.set_index('date', inplace=True)
        df.sort_index(inplace=True)
        return df

    def filter_trading_session(self):
        """Filter data to trading hours (9:15 AM to 3:30 PM)"""
        self.df = self.df.between_time(self.trading_session_start, self.trading_session_end)
        print("Filtered data shape:", self.df.shape)

    def calculate_indicators(self):
        """Calculate technical indicators optimized for 1-minute data"""
        # Shorter periods for intraday
        sma_window = 20    # 20-minute SMA
        rsi_period = 14    # 14-minute RSI
        atr_period = 14    # 14-minute ATR

        # Price Transformations
        self.df['typical_price'] = (self.df['high'] + self.df['low'] + self.df['close']) / 3
        self.df['ohlc4'] = (self.df['open'] + self.df['high'] + self.df['low'] + self.df['close']) / 4

        # Trend Indicators
        self.df['sma_20'] = self.df['close'].rolling(sma_window).mean()
        self.df['ema_12'] = self.df['close'].ewm(span=12).mean()
        self.df['macd'], self.df['macd_signal'], self.df['macd_hist'] = talib.MACD(self.df['close'])

        # Momentum Indicators
        self.df['rsi'] = talib.RSI(self.df['close'], timeperiod=rsi_period)
        self.df['cci'] = talib.CCI(self.df['high'], self.df['low'], self.df['close'], timeperiod=14)

        # Volatility Indicators
        self.df['atr'] = talib.ATR(self.df['high'], self.df['low'], self.df['close'], timeperiod=atr_period)
        self.df['bb_upper'], self.df['bb_middle'], self.df['bb_lower'] = talib.BBANDS(self.df['close'], timeperiod=20)

        # Opening Range Breakout (9:15-9:30 AM)
        opening_range = self.df.between_time('09:15', '09:30')
        self.df['orb_high'] = opening_range['high'].rolling(15).max().ffill()
        self.df['orb_low'] = opening_range['low'].rolling(15).min().ffill()

        # Forward-fill missing values instead of dropping
        self.df.ffill(inplace=True)
        print("Indicators calculated. Remaining data shape:", self.df.shape)

    def generate_signals(self):
        """Generate trading signals"""
        conditions = [
            self.df['close'] > self.df['orb_high'],
            self.df['close'] > self.df['bb_middle'],
            self.df['rsi'].between(40, 70),
            self.df['macd_hist'] > 0
        ]
        self.df['signal'] = np.where(np.all(conditions, axis=0), 1, 0)
        self.df['position'] = self.df['signal'].diff().fillna(0)
        print("Signals generated. Buy signals:", self.df['signal'].sum())

    def run_backtest(self):
        """Run backtest"""
        if self.df.empty:
            print("No data available for backtesting!")
            return pd.DataFrame()

        portfolio = pd.DataFrame(index=self.df.index)
        portfolio['price'] = self.df['close']
        portfolio['position'] = self.df['position']
        portfolio['holdings'] = 0.0
        portfolio['cash'] = self.initial_capital
        portfolio['total'] = self.initial_capital

        for i in range(1, len(portfolio)):
            current_pos = portfolio['position'].iloc[i]
            prev_total = portfolio['total'].iloc[i-1]

            if current_pos != 0:
                # Calculate trade value with slippage
                trade_value = current_pos * prev_total * self.position_size
                executed_price = portfolio['price'].iloc[i] * (1 + self.slippage if current_pos > 0 else 1 - self.slippage)
                
                # Update holdings and cash
                portfolio.loc[portfolio.index[i], 'holdings'] = trade_value / executed_price
                transaction_cost = abs(trade_value) * self.transaction_cost
                portfolio.loc[portfolio.index[i], 'cash'] = prev_total - trade_value - transaction_cost
            else:
                # No trade, carry forward holdings
                portfolio.loc[portfolio.index[i], 'holdings'] = portfolio['holdings'].iloc[i-1]
                portfolio.loc[portfolio.index[i], 'cash'] = portfolio['cash'].iloc[i-1]

            # Update total portfolio value
            portfolio.loc[portfolio.index[i], 'total'] = (
                portfolio['cash'].iloc[i] + 
                (portfolio['holdings'].iloc[i] * portfolio['price'].iloc[i])
            )

        return portfolio

if __name__ == "__main__":
    bt = IntradayBacktester()
    
    print("Filtering Trading Session...")
    bt.filter_trading_session()
    
    print("Calculating Indicators...")
    bt.calculate_indicators()
    
    print("Generating Signals...")
    bt.generate_signals()
    
    print("Running Backtest...")
    results = bt.run_backtest()
    
    print("\nBacktest Complete!")
    if not results.empty:
        print(results.head())
        print("\nFinal Portfolio Value:", results['total'].iloc[-1])
        results['total'].plot(title="Portfolio Performance")
        plt.show()
    else:
        print("No results to display. Check data/indicators.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load 1-minute stock data from Excel file
def load_data(file_name):
    df = pd.read_excel(file_name, engine='openpyxl', header=0)
    df.columns = df.columns.str.strip().str.lower()  # Normalize column names
    print("Column Names:", df.columns)  # Debugging
    
    if 'date' not in df.columns:
        raise KeyError("Column 'date' not found. Available columns: " + ", ".join(df.columns))
    
    df.rename(columns={'date': 'Datetime'}, inplace=True)
    df['Datetime'] = pd.to_datetime(df['Datetime'])
    df.set_index('Datetime', inplace=True)
    return df

# Compute VWAP
def calculate_vwap(df):
    df['Cumulative_TV'] = (df['close'] * df['volume']).cumsum()
    df['Cumulative_Volume'] = df['volume'].cumsum()
    df['VWAP'] = df['Cumulative_TV'] / df['Cumulative_Volume']
    return df

# Compute Moving Averages (for shorter intraday periods)
def moving_average_crossover(df, short_window=5, long_window=15):
    df['SMA_Short'] = df['close'].rolling(window=short_window).mean()
    df['SMA_Long'] = df['close'].rolling(window=long_window).mean()
    return df

# Backtest Strategies with Profit/Loss in Log for Intraday
def backtest_strategy(df, initial_capital=10000, tax_rate=0.15, stop_loss_pct=0.005, take_profit_pct=0.01):
    capital = initial_capital
    position = 0
    trade_log = []
    balance_history = [capital]
    entry_price = 0  # To track the entry price for calculating profit/loss
    
    for i in range(1, len(df)):
        # Moving Average Crossover Signals
        if df['SMA_Short'].iloc[i] > df['SMA_Long'].iloc[i] and df['SMA_Short'].iloc[i-1] <= df['SMA_Long'].iloc[i-1]:
            # Buy signal
            position = capital / df['close'].iloc[i]
            capital = 0
            entry_price = df['close'].iloc[i]
            trade_log.append((df.index[i], 'BUY', df['close'].iloc[i], 0))  # 0 profit/loss at the moment
        elif df['SMA_Short'].iloc[i] < df['SMA_Long'].iloc[i] and df['SMA_Short'].iloc[i-1] >= df['SMA_Long'].iloc[i-1]:
            # Sell signal
            if position > 0:
                capital = position * df['close'].iloc[i]
                profit_loss = capital - (position * entry_price)  # Profit or Loss calculation
                
                # Apply tax on profit (if any)
                tax_deduction = profit_loss * tax_rate if profit_loss > 0 else 0
                profit_after_tax = profit_loss - tax_deduction
                
                # Stop-Loss and Take-Profit Logic
                if profit_loss < -capital * stop_loss_pct:  # Stop-Loss Trigger
                    profit_after_tax = -capital * stop_loss_pct
                elif profit_loss > capital * take_profit_pct:  # Take-Profit Trigger
                    profit_after_tax = capital * take_profit_pct

                capital += profit_after_tax  # Adjust capital after tax and stop/take profit
                position = 0
                trade_log.append((df.index[i], 'SELL', df['close'].iloc[i], profit_after_tax, tax_deduction))
        
        balance_history.append(capital + (position * df['close'].iloc[i]))
    
    final_capital = capital + (position * df['close'].iloc[-1])
    return trade_log, balance_history, final_capital

# Plot Trade Entries and Exits
def plot_trades(df, trade_log):
    buy_signals = [trade for trade in trade_log if trade[1] == 'BUY']
    sell_signals = [trade for trade in trade_log if trade[1] == 'SELL']
    
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['close'], label='Close Price', color='blue')
    plt.scatter([trade[0] for trade in buy_signals], [trade[2] for trade in buy_signals], marker='^', color='green', label='Buy Signal')
    plt.scatter([trade[0] for trade in sell_signals], [trade[2] for trade in sell_signals], marker='v', color='red', label='Sell Signal')
    plt.title('Trade Signals on Price Chart')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

# Plot Account Balance Over Time
def plot_balance(balance_history):
    plt.figure(figsize=(12, 6))
    plt.plot(balance_history, label='Account Balance', color='purple')
    plt.title('Account Balance Over Time')
    plt.xlabel('Trades')
    plt.ylabel('Balance')
    plt.legend()
    plt.show()

# Main Execution for Intraday Trading
file_name = 'data.xlsx'  # Use your 1-minute intraday data file
df = load_data(file_name)
df = calculate_vwap(df)
df = moving_average_crossover(df)

# Perform the backtest
trade_log, balance_history, final_capital = backtest_strategy(df, stop_loss_pct=0.005, take_profit_pct=0.01)

print(f'Initial Capital: $10000')
print(f'Final Capital: ${final_capital:.2f}')
print(f'Total Profit/Loss: ${final_capital - 10000:.2f}')
print(f'Total Trades Executed: {len(trade_log)}')

# Display trade logs with Profit/Loss
print("Trade Log (with Profit/Loss):")
for trade in trade_log:
    print(f"Date: {trade[0]}, Action: {trade[1]}, Price: ${trade[2]:.2f}, Profit/Loss: ${trade[3]:.2f}")

# Plot trade signals and account balance
plot_trades(df, trade_log)
plot_balance(balance_history)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load minute-level stock data from CSV file (optimized for large data)
def load_data(file_name):
    try:
        # Read in chunks for large CSV files
        df = pd.read_csv(file_name, usecols=['date', 'open', 'high', 'low', 'close', 'volume'])
        print("Data loaded successfully!")
    except Exception as e:
        print(f"Error loading data: {e}")
        return None

    # Normalize column names and parse dates
    df.columns = df.columns.str.strip().str.lower()
    df['datetime'] = pd.to_datetime(df['date'])  # Ensure correct datetime format
    df.set_index('datetime', inplace=True)
    df.drop(columns=['date'], inplace=True)  # Remove the 'date' column after setting as index
    
    print("Data preview:")
    print(df.head())  # Print first few rows to inspect data
    
    return df

# Calculate Exponential Moving Averages (EMA)
def calculate_ema(df, short_period=5, long_period=20):
    df['EMA_Short'] = df['close'].ewm(span=short_period, adjust=False).mean()
    df['EMA_Long'] = df['close'].ewm(span=long_period, adjust=False).mean()
    return df

# Calculate Relative Strength Index (RSI)
def calculate_rsi(df, period=14):
    delta = df['close'].diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))
    return df

# Calculate Bollinger Bands
def calculate_bollinger_bands(df, window=20, num_std_dev=2):
    rolling_mean = df['close'].rolling(window=window).mean()
    rolling_std = df['close'].rolling(window=window).std()
    
    df['Bollinger_Upper'] = rolling_mean + (rolling_std * num_std_dev)
    df['Bollinger_Lower'] = rolling_mean - (rolling_std * num_std_dev)
    return df

# Backtest the Scalping Strategy
def backtest_strategy(df, initial_capital=10000, stop_loss_pct=0.005, take_profit_pct=0.01, position_size_pct=0.1):
    capital = initial_capital
    position = 0  # Number of shares held
    entry_price = 0  # Entry price of the current position
    trade_log = []
    balance_history = [capital]  # Track account balance over time
    
    for i in range(1, len(df)):
        current_price = df['close'].iloc[i]
        high = df['high'].iloc[i]
        low = df['low'].iloc[i]
        
        # Check if we have an open position to manage stop-loss/take-profit
        if position > 0:
            stop_loss_price = entry_price * (1 - stop_loss_pct)
            take_profit_price = entry_price * (1 + take_profit_pct)
            exit_reason = None
            exit_price = None
            
            # Check if stop-loss was hit (using low of the minute)
            if low <= stop_loss_price:
                exit_price = stop_loss_price
                exit_reason = 'Stop-Loss'
            # Check if take-profit was hit (using high of the minute)
            elif high >= take_profit_price:
                exit_price = take_profit_price
                exit_reason = 'Take-Profit'
            
            # If exit due to stop-loss or take-profit
            if exit_price is not None:
                capital += position * exit_price
                profit_loss = (exit_price - entry_price) * position
                trade_log.append((df.index[i], 'SELL', exit_price, profit_loss, exit_reason))
                position = 0
                balance_history.append(capital)
                continue  # Position closed, move to next iteration
        
        # Buy Signal: EMA crossover up and RSI not overbought
        if (df['EMA_Short'].iloc[i] > df['EMA_Long'].iloc[i] and 
            df['EMA_Short'].iloc[i-1] <= df['EMA_Long'].iloc[i-1] and 
            df['RSI'].iloc[i] < 70):
            if position == 0:
                # Calculate position size based on current capital
                position_value = capital * position_size_pct
                position = position_value / current_price
                entry_price = current_price
                capital -= position_value  # Deduct the invested amount
                trade_log.append((df.index[i], 'BUY', current_price, 0, 'Signal'))
                balance_history.append(capital + position * current_price)
        
        # Sell Signal: EMA crossover down and RSI not oversold
        elif (df['EMA_Short'].iloc[i] < df['EMA_Long'].iloc[i] and 
              df['EMA_Short'].iloc[i-1] >= df['EMA_Long'].iloc[i-1] and 
              df['RSI'].iloc[i] > 30):
            if position > 0:
                # Sell based on signal at current closing price
                capital += position * current_price
                profit_loss = (current_price - entry_price) * position
                trade_log.append((df.index[i], 'SELL', current_price, profit_loss, 'Signal'))
                position = 0
                balance_history.append(capital)
        
        # Update balance history if position remains open
        if position > 0:
            balance_history.append(capital + position * current_price)
        else:
            balance_history.append(capital)
    
    # Close any remaining position at end of data
    if position > 0:
        capital += position * df['close'].iloc[-1]
        profit_loss = (df['close'].iloc[-1] - entry_price) * position
        trade_log.append((df.index[-1], 'SELL', df['close'].iloc[-1], profit_loss, 'End'))
        position = 0
        balance_history.append(capital)
    
    final_capital = capital
    return trade_log, balance_history, final_capital

# Plot Trade Entries and Exits
def plot_trades(df, trade_log):
    buy_signals = [trade for trade in trade_log if trade[1] == 'BUY']
    sell_signals = [trade for trade in trade_log if trade[1] == 'SELL']
    
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['close'], label='Close Price', color='blue')
    plt.scatter([trade[0] for trade in buy_signals], [trade[2] for trade in buy_signals], marker='^', color='green', label='Buy Signal')
    plt.scatter([trade[0] for trade in sell_signals], [trade[2] for trade in sell_signals], marker='v', color='red', label='Sell Signal')
    plt.title('Scalping Strategy: Trade Signals')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

# Plot Account Balance Over Time
def plot_balance(balance_history):
    plt.figure(figsize=(12, 6))
    plt.plot(balance_history, label='Account Balance', color='purple')
    plt.title('Account Balance Over Time')
    plt.xlabel('Trades')
    plt.ylabel('Balance')
    plt.legend()
    plt.show()

# Main Execution for Scalping Strategy with Intraday Data (1-minute data)
file_name = 'data.csv'  # Use your intraday 1-minute data file
df = load_data(file_name)
df = calculate_ema(df)
df = calculate_rsi(df)
df = calculate_bollinger_bands(df)

# Perform the backtest for scalping strategy
trade_log, balance_history, final_capital = backtest_strategy(df, stop_loss_pct=0.005, take_profit_pct=0.01)

# Print results
print(f'Initial Capital: $10000')
print(f'Final Capital: ${final_capital:.2f}')
print(f'Total Profit/Loss: ${final_capital - 10000:.2f}')
print(f'Total Trades Executed: {len(trade_log)}')

# Display trade logs with Profit/Loss
print("Trade Log (with Profit/Loss):")
for trade in trade_log[:5]:  # Print only the first 5 trades
    print(f"Date: {trade[0]}, Action: {trade[1]}, Price: ${trade[2]:.2f}, Profit/Loss: ${trade[3]:.2f}, Reason: {trade[4]}")
    
if len(trade_log) > 5:
    print("...")
    
for trade in trade_log[-5:]:  # Print only the last 5 trades
    print(f"Date: {trade[0]}, Action: {trade[1]}, Price: ${trade[2]:.2f}, Profit/Loss: ${trade[3]:.2f}, Reason: {trade[4]}")

# Plot trade signals and account balance
plot_trades(df, trade_log)
plot_balance(balance_history)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load minute-level stock data from CSV file
def load_data(file_name):
    df = pd.read_excel(file_name)
    df.columns = df.columns.str.strip().str.lower()  # Normalize column names
    df['datetime'] = pd.to_datetime(df['date'])  # Ensure correct datetime format
    df.set_index('datetime', inplace=True)
    return df

# Calculate Exponential Moving Averages (EMA)
def calculate_ema(df, short_period=5, long_period=20):
    df['EMA_Short'] = df['close'].ewm(span=short_period, adjust=False).mean()
    df['EMA_Long'] = df['close'].ewm(span=long_period, adjust=False).mean()
    return df

# Calculate Relative Strength Index (RSI)
def calculate_rsi(df, period=14):
    delta = df['close'].diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))
    return df

# Calculate Bollinger Bands
def calculate_bollinger_bands(df, window=20, num_std_dev=2):
    rolling_mean = df['close'].rolling(window=window).mean()
    rolling_std = df['close'].rolling(window=window).std()
    
    df['Bollinger_Upper'] = rolling_mean + (rolling_std * num_std_dev)
    df['Bollinger_Lower'] = rolling_mean - (rolling_std * num_std_dev)
    return df

# Backtest the Scalping Strategy
def backtest_strategy(df, initial_capital=10000, stop_loss_pct=0.005, take_profit_pct=0.01, position_size_pct=0.1):
    capital = initial_capital
    position = 0
    entry_price = 0
    trade_log = []
    balance_history = [capital]
    
    for i in range(1, len(df)):
        # Buy Signal: 5-period EMA crosses above 20-period EMA and RSI < 70
        if df['EMA_Short'].iloc[i] > df['EMA_Long'].iloc[i] and df['EMA_Short'].iloc[i-1] <= df['EMA_Long'].iloc[i-1] and df['RSI'].iloc[i] < 70:
            if position == 0:  # Only buy if no position is open
                position = capital * position_size_pct / df['close'].iloc[i]
                capital -= position * df['close'].iloc[i]
                entry_price = df['close'].iloc[i]
                trade_log.append((df.index[i], 'BUY', df['close'].iloc[i], 0))
        
        # Sell Signal: 5-period EMA crosses below 20-period EMA and RSI > 30
        elif df['EMA_Short'].iloc[i] < df['EMA_Long'].iloc[i] and df['EMA_Short'].iloc[i-1] >= df['EMA_Long'].iloc[i-1] and df['RSI'].iloc[i] > 30:
            if position > 0:  # Only sell if a position is open
                capital += position * df['close'].iloc[i]
                profit_loss = capital - (position * entry_price)
                
                # Apply stop-loss or take-profit
                if profit_loss < -capital * stop_loss_pct:
                    profit_loss = -capital * stop_loss_pct
                elif profit_loss > capital * take_profit_pct:
                    profit_loss = capital * take_profit_pct
                
                capital += profit_loss
                position = 0
                trade_log.append((df.index[i], 'SELL', df['close'].iloc[i], profit_loss))
        
        # Record the balance history
        balance_history.append(capital + (position * df['close'].iloc[i]))
    
    final_capital = capital + (position * df['close'].iloc[-1])
    return trade_log, balance_history, final_capital

# Plot Trade Entries and Exits
def plot_trades(df, trade_log):
    buy_signals = [trade for trade in trade_log if trade[1] == 'BUY']
    sell_signals = [trade for trade in trade_log if trade[1] == 'SELL']
    
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['close'], label='Close Price', color='blue')
    plt.scatter([trade[0] for trade in buy_signals], [trade[2] for trade in buy_signals], marker='^', color='green', label='Buy Signal')
    plt.scatter([trade[0] for trade in sell_signals], [trade[2] for trade in sell_signals], marker='v', color='red', label='Sell Signal')
    plt.title('Scalping Strategy: Trade Signals')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

# Plot Account Balance Over Time
def plot_balance(balance_history):
    plt.figure(figsize=(12, 6))
    plt.plot(balance_history, label='Account Balance', color='purple')
    plt.title('Account Balance Over Time')
    plt.xlabel('Trades')
    plt.ylabel('Balance')
    plt.legend()
    plt.show()

# Main Execution for Scalping Strategy with Intraday Data (1-minute data)
file_name = 'data.xlsx'  # Use your intraday 1-minute data file
df = load_data(file_name)
df = calculate_ema(df)
df = calculate_rsi(df)
df = calculate_bollinger_bands(df)

# Perform the backtest for scalping strategy
trade_log, balance_history, final_capital = backtest_strategy(df, stop_loss_pct=0.005, take_profit_pct=0.01)

# Print results
print(f'Initial Capital: $10000')
print(f'Final Capital: ${final_capital:.2f}')
print(f'Total Profit/Loss: ${final_capital - 10000:.2f}')
print(f'Total Trades Executed: {len(trade_log)}')

# Display trade logs with Profit/Loss
print("Trade Log (with Profit/Loss):")
for trade in trade_log:
    print(f"Date: {trade[0]}, Action: {trade[1]}, Price: ${trade[2]:.2f}, Profit/Loss: ${trade[3]:.2f}")

# Plot trade signals and account balance
plot_trades(df, trade_log)
plot_balance(balance_history)
```
